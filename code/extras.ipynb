```python
GHInventory_df = pd.read_csv('../data/01_input/00_InventoryGas.csv', encoding='utf-8')
GHInventory_df.head(5)
```

```python
GHInventory_df.info()
```

```python
GHInventory_df['category'].value_counts()
```

```python
# simplify column names
GHInventory_df['category'] = np.where(GHInventory_df['category'] == "methane_ch4_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent", 
    'methane_ch4', np.where(GHInventory_df['category'] == "carbon_dioxide_co2_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent", 
    'carbon_dioxide_co2', np.where(GHInventory_df['category'] == "greenhouse_gas_ghgs_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent", 
    'greenhouse_gas_ghgs', np.where(GHInventory_df['category'] == "nitrous_oxide_n2o_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent", 
    'nitrous_oxide_n2o', np.where(GHInventory_df['category'] == "sulphur_hexafluoride_sf6_emissions_in_kilotonne_co2_equivalent", 
    'sulphur_hexafluoride_sf6', np.where(GHInventory_df['category'] == "hydrofluorocarbons_hfcs_emissions_in_kilotonne_co2_equivalent", 
    'hydrofluorocarbons_hfcs', np.where(GHInventory_df['category'] == "greenhouse_gas_ghgs_emissions_including_indirect_co2_without_lulucf_in_kilotonne_co2_equivalent", 
    'greenhouse_gas_ghgs_indirect', np.where(GHInventory_df['category'] == "perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent", 
    'perfluorocarbons_pfcs', np.where(GHInventory_df['category'] == "nitrogen_trifluoride_nf3_emissions_in_kilotonne_co2_equivalent", 
    'nitrogen_trifluoride_nf3', np.where(GHInventory_df['category'] == "unspecified_mix_of_hydrofluorocarbons_hfcs_and_perfluorocarbons_pfcs_emissions_in_kilotonne_co2_equivalent", 
    'hydrofluorocarbons_hfcs', GHInventory_df['category']))))))))))

GHInventory_df
                            
```

```python
#add indicators as columns instead of rows
GHInventory_df = GHInventory_df.pivot_table(values='value', index=['country_or_area','year'], 
                                            columns='category').reset_index().rename_axis(None, axis=1)
GHInventory_df
```

```python
GHInventory_df['country_or_area'].value_counts()
```

```python
GHInventory_df = GHInventory_df[GHInventory_df['year'] >= 2012]
```

```python
GHInventory_df['year'].value_counts()
```

```python
GHInventory_df = GHInventory_df.rename(
    columns={"country_or_area": "Country Name",
            "year": "Year"})
```

```python
#add country code
GHInventory_df['Country Code'] = GHInventory_df['Country Name'].apply(lambda x: do_fuzzy_search(x))
GHInventory_df
```

```python
GHInventory_df = GHInventory_df[~GHInventory_df['Country Name'].str.contains("European Union")]
```

```python
GHInventory_df[GHInventory_df['Country Code'].isna()]['Country Name'].value_counts()
```

```python
#reorder columns to add prefix
Column_names = Ind + GHInventory_df.columns[~GHInventory_df.columns.isin(Ind)].tolist()
GHInventory_df = GHInventory_df[Column_names]
GHInventory_df.columns
```

```python
#rename not identification columns
newnames = "GHI_" + GHInventory_df.columns[~GHInventory_df.columns.isin(Ind)]
GHInventory_df.columns = Ind + newnames.tolist()
```
