Code for an interactive world map that displays city emissions data over the years. 

# Packages 

```python
#Load packages 
import plotly.express as px
import pandas as pd
```

# *Data Preprocessing*

```python
#Load data
#df = pd.read_excel("https://github.com/azunig/kul-mda-a1/blob/main/data/02_output/02_DataToDescriptive_coordinates.xlsx")
#Can't load data from the github as it is private, will load from local disk for now

df = pd.read_excel("https://github.com/jadenio/CityEmissionsData/blob/main/final_df.xlsx?raw=true")
```

```python
#Sort by year
df = df.sort_values('Year', ascending=True)
```

# Create Map

```python
#Create map figure 
map_emissions = px.scatter_geo(
    df, #emissions data
    lat = df['Latitude'], #latitude data
    lon = df['Longitude'], #longitude data
    #zoom = 0.5, #Starting zoom level
    color = 'TotalEmissions', #Colored by Emissions per capita  
    color_continuous_scale='RdYlGn_r', #Color scale for Emissions per capita
    size = 'TotalEmissions', #size of markers based on Emissions
    size_max = 50, #Create scale for marker sizes
    opacity = .6,
    animation_frame = 'Year', #slider values
    #animation_group = 'City', #Slider groups
    hover_name = 'City',
    hover_data = ['TotalEmissions','Year','Population'], #data to be displayed when mouse is hovered over point
    #mapbox_style="carto-positron", #for flat map
    projection = 'natural earth'
)
```
